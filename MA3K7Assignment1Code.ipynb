```python
# x = 0,...,9 is the number we choose to determine whether (x, y) is a pair of numbers that has been used already
x = 0

# computing the first bracelet given in the question
a = 1
b = 5

arrs = []
arrs.append(a)
arrs.append(b)

for i in range(0,100):
    # addition modulo 10, generating the next number in the bracelet
    c = (a+b)%10
    a = b
    b = c
    arrs.append(c)
    # checking if the pairs (x,b) have been used yet
    if a == x:
        print((x,b), "has been used")
    # code stops when the bracelet 'joins up', prints the bracelet array and length    
    if a == 1 and b == 5:
        print("bracelet 1:", arrs[:-2])
        print("The length of this bracelet is:", len(arrs)-2)
        break

# computing the second bracelet
a = 0
b = 0
arrs2 = []
arrs2.append(a)
arrs2.append(b)
for i in range(0,100):
    c = (a+b)%10
    a = b
    b = c
    arrs2.append(c)
    
    if a == x:
        print((x,b), "has been used")
        
    if a == 0 and b == 0:
        print("bracelet 2:", arrs2[:-2])
        print("The length of this bracelet is:", len(arrs2)-2)
        break
        
# computing the third bracelet
a = 0
b = 2
arrs3 = []
arrs3.append(a)
arrs3.append(b)
for i in range(0,100):
    c = (a+b)%10
    a = b
    b = c
    arrs3.append(c)
    
    if a == x:
        print((x,b), "has been used")
        
    if a == 0 and b == 2:
        print("bracelet 3:", arrs3[:-2])
        print("The length of this bracelet is:", len(arrs3)-2)
        break

# computing the fourth bracelet
a = 0
b = 5
arrs4 = []
arrs4.append(a)
arrs4.append(b)
for i in range(0,100):
    c = (a+b)%10
    a = b
    b = c
    arrs4.append(c)
    
    if a == x:
        print((x,b), "has been used")
        
    if a == 0 and b == 5:
        print("bracelet 4:", arrs4[:-2])
        print("The length of this bracelet is:", len(arrs4)-2)
        break
        
# computing the fifth bracelet
a = 1
b = 3
arrs5 = []
arrs5.append(a)
arrs5.append(b)
for i in range(0,100):
    c = (a+b)%10
    a = b
    b = c
    arrs5.append(c)
    
    if a == x:
        print((x,b), "has been used")
        
    if a == 1 and b == 3:
        print("bracelet 5:", arrs5[:-2])
        print("The length of this bracelet is:", len(arrs5)-2)
        break
        
# computing the sixth bracelet
a = 2
b = 6
arrs6 = []
arrs6.append(a)
arrs6.append(b)
for i in range(0,100):
    c = (a+b)%10
    a = b
    b = c
    arrs6.append(c)
    
    if a == x:
        print((x,b), "has been used")
        
    if a == 2 and b == 6:
        print("bracelet 6:", arrs6[:-2])
        print("The length of this bracelet is:", len(arrs6)-2)
        break
```
