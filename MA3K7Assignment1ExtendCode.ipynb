```python
# x = 0,...,6 is the number we choose to determine whether (x, y) is a pair of numbers that has been used already
x = 0

# computing the first bracelet from a random starting pair of numbers
a = 1
b = 6

arrs = []
arrs.append(a)
arrs.append(b)

for i in range(0,100):
    # addition modulo 7, generating the next number in the bracelet
    c = (a+b)%7
    a = b
    b = c
    arrs.append(c)
    # checking if the pairs (x,b) have been used yet
    if a == x:
        print((x,b), "has been used")
    # code stops when the bracelet 'joins up', prints the bracelet array and length    
    if a == 1 and b == 6:
        print("bracelet 1:", arrs[:-2])
        print("The length of this bracelet is:", len(arrs)-2)
        break

# computing the second bracelet
a = 0
b = 0
arrs2 = []
arrs2.append(a)
arrs2.append(b)
for i in range(0,100):
    c = (a+b)%7
    a = b
    b = c
    arrs2.append(c)
    
    if a == x:
        print((x,b), "has been used")
        
    if a == 0 and b == 0:
        print("bracelet 2:", arrs2[:-2])
        print("The length of this bracelet is:", len(arrs2)-2)
        break
        
# computing the third bracelet
a = 0
b = 2
arrs3 = []
arrs3.append(a)
arrs3.append(b)
for i in range(0,100):
    c = (a+b)%7
    a = b
    b = c
    arrs3.append(c)
    
    if a == x:
        print((x,b), "has been used")
        
    if a == 0 and b == 2:
        print("bracelet 3:", arrs3[:-2])
        print("The length of this bracelet is:", len(arrs3)-2)
        break
        
# computing the fourth bracelet
a = 0
b = 4
arrs4 = []
arrs4.append(a)
arrs4.append(b)
for i in range(0,100):
    c = (a+b)%7
    a = b
    b = c
    arrs4.append(c)
    
    if a == x:
        print((x,b), "has been used")
        
    if a == 0 and b == 4:
        print("bracelet 4:", arrs4[:-2])
        print("The length of this bracelet is:", len(arrs4)-2)
        break
```
